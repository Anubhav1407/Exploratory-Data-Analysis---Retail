```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(),annot=True)
plt.title("HEAT MAP")
plt.show()
```

```python
df.corr()
```

```python
df["Ship Mode"].value_counts()
```

```python
sns.countplot(x=df["Ship Mode"])
```

```python
sns.pairplot(df)
```

```python
ps = df.groupby("Ship Mode")[["Profit", "Sales"]].sum()
ps
```

```python
ps.plot(kind="bar", figsize=(10,6))
plt.title("Profit v/s Sales")
plt.ylabel("Profit and Sales")
plt.show()
```

```python
mp = df.groupby("Category")[["Profit", "Sales"]].sum()
mp
```

```python
mp.plot(kind="bar", figsize=(10,6))
plt.title("Profit v/s Sales")
plt.ylabel("Profit and Sales")
plt.show()
```

```python
cat = df["Category"].value_counts()
cat
```

```python
explode_list=[0, 0.1, 0.1]
colors_list=["lightcoral", "pink", "lightblue"]
cat.plot(kind="pie", figsize=(15,6), autopct="%1.1f%%", shadow=True, startangle=180, explode=explode_list, labels=None, pctdistance=1.12, colors=colors_list)

plt.title("Category")
plt.axis("equal")
plt.legend(labels=cat.index, loc="upper left")
plt.show()
```

```python
sub = df["Sub-Category"].value_counts()
sub
```

```python
np = df.groupby("Sub-Category")["Sales"].sum()
np
```

```python
np.plot(kind="barh", figsize=(10,6))
plt.title("Sub-Category v/s Sales")
plt.ylabel("Sales")
plt.show()
```

```python
tp = df.groupby("Sub-Category")["Profit"].sum()
tp
```

```python
tp.plot(kind="barh", figsize=(10,6))
plt.title("Sub-Category v/s Profit")
plt.ylabel("Sub-Categories")
plt.xlabel("Profit")
plt.show()
```

```python
state = df["State"].value_counts()
state
```

```python
state.plot(kind="bar",figsize=(15,6))
plt.title("Distribution of Buyers in each State")
plt.xlabel("States")
plt.ylabel("Number of Buyers")
plt.show()
```

```python
ss=df.groupby(["State"]).Sales.sum()
sp=df.groupby(["State"]).Profit.sum()
```

```python
ss.sort_values(ascending=False).plot(kind="bar",figsize=(15,6))
plt.xlabel("States")
plt.ylabel("Sales")
plt.show()
```

```python
sp.sort_values(ascending=False).plot(kind="bar",figsize=(15,6))
plt.xlabel("States")
plt.ylabel("Profit")
plt.show()
```

```python
segment=df["Segment"].value_counts()
segment
```

```python
segment.plot(kind="pie",figsize=(7,7),autopct="%1.1f%%",shadow=True, labels=None, colors=["lightcoral","pink","lightblue"])
plt.title("Segment")
plt.legend(labels=segment.index, loc="upper left")
plt.axis("equal")
plt.show()
```

```python
df.groupby("Segment")[["Sales","Profit"]].sum().plot(kind="bar", color=["lightskyblue","lightcoral"])
plt.ylabel("Sales and Profit")
plt.xlabel("Segment")
plt.title("Sales and Profit distribution among segments")
plt.show()
```

```python
reg=df["Region"].value_counts()
reg
```

```python
reg.plot(kind="pie",figsize=(7,7),autopct="%1.1f%%",shadow=True, labels=None, colors=["lightcoral","pink","lightblue", "lightsalmon"])
plt.title("Region")
plt.legend(labels=reg.index, loc="upper left")
plt.axis("equal")
plt.show()
```

```python
df.groupby("Region")[["Sales","Profit"]].sum().sort_values(by=["Sales","Profit"],ascending=False).plot(kind="bar", color=["lightskyblue","lightcoral"])
plt.ylabel("Sales and Profit")
plt.xlabel("Region")
plt.title("Sales and Profit distribution among regions")
plt.show()
```
